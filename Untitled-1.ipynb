```python
from sodapy import Socrata
import pandas as pd
cliente = Socrata("www.datos.gov.co", None)
resultado = cliente.get("8cgj-t5ds",limit=10100)
Dframe=pd.DataFrame.from_records(resultado)
Dframe.head(20)     
```

```python
Dframe.columns
```

```python
Dframe['n_mero_acumulado_de_1_dosis']=Dframe['n_mero_acumulado_de_1_dosis'].astype(int)
Dframe['n_mero_acumulado_de_esquema']=Dframe['n_mero_acumulado_de_esquema'].astype(int)
```

```python
Dframe.dtypes
```

```python
Dframe=Dframe.drop(Dframe[(Dframe['grupo_edad']=='No definido')].index)
```

```python
Dframe=Dframe.drop(Dframe[Dframe['grupo_edad']=='Todas'].index)
```

```python
df = Dframe.groupby(['entidad_territorial'],as_index=False).agg({
            'n_mero_acumulado_de_1_dosis':'sum',
            'n_mero_acumulado_de_esquema':'sum'
    })
df2=df.sort_values('entidad_territorial')
df2    
```

```python
df = Dframe.groupby(['grupo_edad'],as_index=False).agg({
            'n_mero_acumulado_de_1_dosis':'sum',
            'n_mero_acumulado_de_esquema':'sum'
    })
df2=df.sort_values('grupo_edad')
df2    
```

```python
df = Dframe.groupby(['entidad_territorial','municipio'],as_index=False).agg({
            'n_mero_acumulado_de_1_dosis':'sum',
            'n_mero_acumulado_de_esquema':'sum'
    })
df2=df.sort_values('municipio')
df2.head(20) 
df2[df['entidad_territorial']=='Vaupés']  
```
